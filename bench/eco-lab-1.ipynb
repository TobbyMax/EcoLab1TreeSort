```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('results.csv', sep=',')
data
```

```python
# Группировка данных по типам данных
grouped_data = data.groupby('type')

# Создание сетки из 4 графиков
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Построение графиков для каждого типа данных
for (name, group), ax in zip(grouped_data, axs.flatten()):
    for sort_type in group['algorithm'].unique():
        group_sort = group[group['algorithm'] == sort_type]
        ax.plot(group_sort['size'], group_sort['time'], label=sort_type)
    ax.set_title(f'Зависимость времени сортировки от размера массива для типа {name}')
    ax.set_xlabel('Размер массива')
    ax.set_ylabel('Время сортировки')
    ax.legend()
    ax.set_yscale('log') # Установка логарифмической шкалы для оси Y

plt.tight_layout()
plt.show()

```
